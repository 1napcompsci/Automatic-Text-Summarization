```python
import pandas as pd
import re
import jieba
```

```python
news100 = pd.read_excel('appledaily_politinews100.xlsx')
# 文章3，文章29，和文章76含韓國瑜的新聞
koreaLst = [news100['content'][3], news100['content'][29], news100['content'][76]]
```

```python
# 選擇文章，把文章分成句子
def splitSentences(document):
    #for sent in re.findall(u'[^!?。\.\!\?]+[!?。\.\!\?]?', paragraph, flags=re.U):
    for sent in re.findall(u'[^!?。\!\?]+[!?。\!\?]?', document, flags=re.U):
        yield sent

corpus = []
for doc in koreaLst:
    for sent in list(splitSentences(doc))[:-1]:
        corpus.append(' '.join(jieba.cut(sent)))
corpus
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
vect = vectorizer.fit_transform(corpus)
```

```python
vect.toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cs = cosine_similarity(vect)
```

```python
cs.shape
```

```python
from sklearn import cluster
km = cluster.KMeans(n_clusters=3, init='k-means++', random_state=42)
c = km.fit_predict(cs)
```

```python
import pandas as pd
import re
import jieba
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score
import numpy as np
```

```python
news100 = pd.read_excel('appledaily_politinews100.xlsx')
news100.head()
```

```python
# 選擇文章，把文章分成句子
docNum = 31
testDoc = news100['content'][docNum]
testTitle = news100['title'][docNum]

def splitSentences(document):
    #for sent in re.findall(u'[^!?。\.\!\?]+[!?。\.\!\?]?', paragraph, flags=re.U):
    for sent in re.findall(u'[^!?。\!\?]+[!?。\!\?]?', document, flags=re.U):
        yield sent

sentenceDoc = []
for sent in list(splitSentences(testDoc))[:-1]:
    sentenceDoc.append(' '.join(jieba.cut(sent)))

# 原本文章
sentenceDoc
```

```python
vect = CountVectorizer()
X = vect.fit_transform(sentenceDoc)
```

```python
true_k = 3
model = KMeans(n_clusters=true_k, init='k-means++', max_iter=100, n_init=1)
model.fit(X)
```

```python
print("Top terms per cluster:")
order_centroids = model.cluster_centers_.argsort()[:, ::-1]
terms = vect.get_feature_names()
for i in range(true_k):
    print("Cluster %d:" % i,)
    for ind in order_centroids[i, :10]:
        print(' %s' % terms[ind],)
    print()
```
